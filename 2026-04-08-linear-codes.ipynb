# Linear codes in SageMath

Creating the alphabet, that is, the finite field in which you want to work:

```python
F = GF(2) #the binary field
```

Showing the object we just created, for inspection

```python
F
```

Creating a generator matrix for the code:

```python
G = Matrix(F, [
    [1, 0, 1],
    [0, 1, 1]
])
```

Putting the matrix in its systematic form (if needed):

```python
G_sys = G.echelon_form()
```

Construct the code $C$:

```python
C = LinearCode(G)
```

Compute the parity-check matrix $H$:

```python
H = C.parity_check_matrix()
H
```

Compute the parameters $n,k,d$ of the code:

```python
n = C.length()
k = C.dimension()
d = C.minimum_distance()
print(f"Parameters: [{n},{k},{d}]")
```

Encode a message $u$. The message must be an element of $\mathbb{F}_q^k$.

```python
u = vector(F, [1, 0])
x = C.encode(u)
x
```

Decode a received message:

```python
y = vector(F, [0, 0, 1]) #Simulate a received message with an error
C.decode_to_code(y) # Decode the received word to the nearest codeword.
```

```python
C.decode_to_message(y) # Decode the received word to the original message
```

Compute the dual of the code $C$:

```python
C_dual = C.dual_code()
```

Check self-orthogonality, that is, if $C\subseteq C^\perp$, and self-duality, that is, if $C = C^\perp$:

```python
C.is_self_orthogonal()
```

```python
C.is_self_dual()
```
