We now turn to AG code, and construct the AG code we saw during the lecture from the elliptic curve $E:y^2=x^3-3x$ over $\mathbb{F}_5$, the divisor $D=5\mathcal{O}$ and all rational points on $E$ but $\mathcal{O}$ as evaluation points.

```python
F=GF(5)
P.<x,y,z> = ProjectiveSpace(F, 2) #Build the projective space
E = Curve(y^2*z - x^3-3*x*z^2) #Build the curve
F = E.function_field() #Build the function field F
points = F.places() #Build the set of rational points, seen as places in F
O = E([0,1,0]) #Set O to be the point at infinity
Z, = O.places() #See O as place in F
points.remove(Z) #Remove O from the set of points
D=5*Z #Build the divisor
D.basis_function_space() #Compute a basis of the Riemann-Roch space of D
```

```python
AG=codes.EvaluationAGCode(points, D)
AG
```
